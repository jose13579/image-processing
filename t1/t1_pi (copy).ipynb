```python
from scipy import misc
from scipy import ndimage
import numpy as np
import imageio
import matplotlib.pyplot as plt
import cv2
# open image file and stores it in a numpy array
im = cv2.imread("house.png",0)
plt.imshow(im, cmap=plt.cm.gray)
plt.xlabel('Imagen original')
plt.show()

plt.hist(im.ravel(),256,[0,256]);
plt.xlabel("Niveis de Cinza")
plt.ylabel("Frequencia")
plt.show()

# print image dimensions and max, min mean intensity
largura = im.shape[0]
altura = im.shape[1]
max_intensity = im.max()
min_intensity = im.min()
mean_intensity = im.mean()

print "largura: ", largura
print "altura: ", altura
print "intensidade mínima: ", min_intensity
print "intensidade máxima: ", max_intensity
print "intensidade média: ", mean_intensity

im_negative = 255 - im

plt.imshow(im_negative, cmap=plt.cm.gray)
plt.xlabel('Imagen negativa')
plt.show()

#transformacao das intensidades originais
im_nova = np.zeros(shape=(im.shape[0],im.shape[1]))
im_nova = im*(float(180-120)/float(im.max()-im.min()))+120
        
plt.imshow(im_nova, cmap=plt.cm.gray)
plt.xlabel('Imagen transformada nivel de intensidad')
plt.show()
```

```python
from skimage import io
img = io.imread('city.png')
plt.imshow(img, cmap=plt.cm.gray)
plt.title("Imagem original")
plt.show()

plt.hist(img.ravel(),256,[0,256]); 
plt.xlabel("Niveis de Cinza")
plt.ylabel("Frequencia")
plt.show()

from skimage import util
inverted_img = util.invert(img)

plt.imshow(inverted_img)
plt.title("Imagem original")
plt.show()

from skimage import exposure
better_contrast = exposure.rescale_intensity(img)
img2 = exposure.rescale_intensity(img, out_range=(210, 251))

plt.imshow(img2)
plt.title("Imagem original")
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy import misc
import imageio
import copy

img = misc.imread('baboon.png')
plt.imshow(img, cmap=plt.cm.gray)
plt.xlabel("Imagem original")
plt.show()

plt.hist(img.ravel(),256,[0,256]);
plt.xlabel("Niveis de Cinza")
plt.ylabel("Frequencia")
plt.show()

largura = img.shape[0]
altura = img.shape[1]
max_intensity = img.max()
min_intensity = img.min()
mean_intensity = img.mean()

print "largura: ", largura
print "altura: ", altura
print "intensidade mínima: ", min_intensity
print "intensidade máxima: ", max_intensity
print "intensidade média: ", mean_intensity


img_inv = 255-img
plt.imshow(img_inv, cmap=plt.cm.gray)
plt.xlabel("Imagem invertida")
plt.show()

img_trans = copy.copy(img)
img_trans = img_trans*(float(180-120)/float(img_trans.max()-img_trans.min()))+120
plt.imshow(img_trans, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.xlabel('Imagen transformada nivel de intensidad')
plt.show()
```
