```python
from scipy import misc
from scipy import ndimage
import numpy as np
import imageio
import matplotlib.pyplot as plt
import cv2
# open image file and stores it in a numpy array
im = cv2.imread("baboon.png",0)
plt.imshow(im)
plt.title("Imagem original")
plt.show()

plt.hist(im.ravel(),256,[0,256]); 
plt.xlabel("Niveis de Cinza")
plt.ylabel("Frequencia")
plt.show()
```

```python
# print image dimensions and max, min mean intensity
n, _, _ = plt.hist(im.ravel(),256,[0,256]);

largura = im.shape[0]
altura = im.shape[1]
max_intensity = np.max(np.where(n != 0))
min_intensity = np.min(np.where(n != 0))

print "largura: ", largura
print "altura: ", altura
print "intensidade minima: ", min_intensity
print "intensidade máxima: ", max_intensity
```

```python
im_negative = 255 - im

plt.imshow(im_negative)
plt.title("Imagem negativa")
plt.show()
```
