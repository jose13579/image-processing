```python
#Pacotes
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Abrir a imagem de entrada e mostrar usando o pacote matplotlib
im = cv2.imread("baboon.png",0)
plt.imshow(im, cmap=plt.cm.gray)
plt.xlabel('Imagen original')
plt.show()

# Mostrar o histograma da imagem de entrada
plt.hist(im.ravel(),256,[0,256]);
plt.xlabel("Niveis de Cinza")
plt.ylabel("Frequencia")
plt.show()
```

```python
# Imprimir as dimençoes da imagem, e a maxima, minima e media intensidade
largura = im.shape[0]
altura = im.shape[1]
max_intensity = im.max() #retorna a maxima intensidade da imagem de entrada
min_intensity = im.min() #retorna a minima intensidade da imagem de entrada
mean_intensity = im.mean() #retorna a intensidade media 
                           #(soma de todas as intensidades dividido entre o numero de intensidades)da imagem de entrada

print "largura: ", largura
print "altura: ", altura
print "intensidade mínima: ", min_intensity
print "intensidade máxima: ", max_intensity
print "intensidade média: ", mean_intensity
```

```python
# Retorna a imagem negativa da imagem de entrada
im_negative = 255 - im #255 - imagem de entrada = imagem negativa da entrada
plt.imshow(im_negative, cmap=plt.cm.gray)
plt.xlabel('Imagen negativa')
plt.show()
```

```python
#Transformacao das intensidades originais a um rango [120,180]
im_nova = np.zeros(shape=(im.shape[0],im.shape[1])) #cria uma nova imagen com valor 0
im_nova = im*(float(180-120)/float(im.max()-im.min()))+120 #transforma o rango de intensidades originais [0,255] para o rango [120,180]
plt.imshow(im_nova, cmap=plt.cm.gray, vmin=0, vmax=255)
plt.xlabel('Imagen transformada nivel de intensidad')
plt.show()
```
